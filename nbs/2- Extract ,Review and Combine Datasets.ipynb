```python
SPAM_CLASSIFIER_DIR = DATASET_DIR / "spam-classifier"
SMS_SPAM_DIR = SPAM_CLASSIFIER_DIR / "spam-sms"
YT_SPAM_DIR = SPAM_CLASSIFIER_DIR / "spam-yt"

SMS_SPAM_DIR.mkdir(exist_ok=True, parents=True)
YT_SPAM_DIR.mkdir(exist_ok=True, parents=True)
```

```python
import pandas as pd
```

```python
for path in SMS_SPAM_DIR.glob('*'):
    print(path)
    
    
sms_spam_input_path = SMS_SPAM_DIR / "SMSSpamCollection" #tsv -> tab seperated values type of file 


#sms data frame
sms_df = pd.read_csv(sms_spam_input_path, sep='\t', header=None)
sms_df.columns = ["LABEL", "TEXT"]
sms_df["SOURCE"] = "sms-spam"
sms_df.head()
```

```python
dfs = []

for path in YT_SPAM_DIR.glob('*.csv'):
    #print(path)
    raw_df= pd.read_csv(path)
    raw_df.rename(columns={"CLASS":"RAW_LABEL", "CONTENT": "TEXT"}, inplace =True)
    raw_df['LABEL'] = raw_df['RAW_LABEL'].apply(lambda x: "spam" if x == 1 else "ham")
    raw_df["SOURCE"] = "youtube-spam"
    
    df =raw_df.copy()[['LABEL', "TEXT", "SOURCE"]]
    dfs.append(df)
    
yt_dfs = pd.concat(dfs)
yt_dfs
```

```python
#combine youtube spam ds and sms spam ds
df = pd.concat([yt_dfs, sms_df])
df
```

```python
#Exporting to csv file
df.to_csv(SPAM_DATASET_PATH, index=)
```
