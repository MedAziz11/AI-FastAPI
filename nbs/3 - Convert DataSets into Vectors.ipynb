```python
import pathlib
import pandas as pd
import random

BASE_DIR = pathlib.Path().resolve().parent
DATASET_DIR = BASE_DIR / "datasets"
ZIP_DIR = DATASET_DIR / "zips"
EXPORT_DIR= DATASET_DIR / "exports"

EXPORT_DIR.mkdir(exist_ok=True, parents=True)

SPAM_DATASET_PATH = EXPORT_DIR / "spam-dataset.csv"
```

```python
df = pd.read_csv(SPAM_DATASET_PATH)
df.head()
```

```python
labels = df['LABEL'].tolist()
texts = df['TEXT'].tolist()
```

```python
label_legend = {"ham":0, "spam":1}
label_legend_int = {v: k for k, v in label_legend.items()}
```

```python
labels_as_int = [label_legend[label] for label in labels]
```

```python
random_idx = random.randint(0, len(labels))
random_idx
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
MAX_NUM_WORDS = 280
```

```python
tokenizer = Tokenizer(num_words=MAX_NUM_WORDS)
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
# so basically we taking eachtext from text and we indexing each word of it with our tokenizer.index_word
sequences
```

```python
words_index = tokenizer.index_word
words_index
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
MAX_SQ_LEN = 300
```

```python
X= pad_sequences(sequences, maxlen=MAX_SQ_LEN)
#to make all sequences have the same length
```

```python
labels_as_int_array = np.asarray(labels_as_int)
```
